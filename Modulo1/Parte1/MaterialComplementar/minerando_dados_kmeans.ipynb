```python
# Importando as bibliotecas
import pandas as pd

# Lendo o conjunto de dados iris
iris = pd.read_csv("dados/iris.csv")
iris.head()
```

```python
# Selecionando apenas as colunas numéricas (4 colunas), sem a coluna categórica (classe pré-definida)
X = iris.iloc[:, 0:4].values
X
```

```python
# Importando a biblioteca para executar o K-Means (ou K-Médias)
from sklearn.cluster import KMeans
```

```python
# Executando o algoritmo com 3 clusters e com o métodos de seleção de centroides aleatório
kmeans = KMeans(n_clusters=3, init='random')
```

```python
# Ajustando o modelo conforme o conjunto de dados
kmeans.fit(X)
```

```python
# Apresentando o centros resultantes pelo algoritmo (3 pontos com o valor em cada variável)
kmeans.cluster_centers_
```

```python
# Ajustando novamente o modelo
kmeans.fit(X)
```

```python
# Apresentando os novos centros resultantes (o algoritmo está operando de forma aleatória)
kmeans.cluster_centers_
```

```python
# Apresentando as distâncias de cada ponto para cada centroide
distance = kmeans.fit_transform(X)
distance
```

```python
# Apresentando o cluster que cada ponto pertence (clusters iniciando pelo valor 0 - 1º grupo)
labels = kmeans.labels_
labels
```

```python
wcss = []
num_min_clusters = 1
num_max_clusters = 20

# Calculando o valor de wcss (soma dos quadrados das distâncias de cada ponto para cada cluster definido pelo algoritmo) para todo o intervalo de clusters
for i in range(num_min_clusters, num_max_clusters):
    kmeans = KMeans(n_clusters=i, init ='random')
    kmeans.fit(X)
    # Apresentando a soma para cada número de clusters
    print(i, kmeans.inertia_)
    wcss.append(kmeans.inertia_)

# Importando a biblioteca gráfica
from matplotlib import pyplot as plt

# Apresentando o gráfico de linha e dispersão do método de elbow
plt.plot(range(num_min_clusters, num_max_clusters), wcss)
plt.scatter(range(num_min_clusters, num_max_clusters), wcss)
plt.title('O Método Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

```python
# Criando o K-means novamente com 3 clusters e definição de centroides aleatórias e ajustando com os pontos antigos
kmeans = KMeans(n_clusters=3, init='random')
kmeans.fit(X)

# Criando novos pontos
data = [
    [4.12, 3.4, 1.6, 0.7],
    [5.2, 5.8, 5.2, 6.7],
    [3.1, 3.5, 3.3, 3.0]
]

# Executando a predição do algoritmo para atribuir a um dos 3 clusters
kmeans.predict(data)
```

```python
# Apresentando um gráfico com os pontos indicados para cada cluster e os centroides
plt.scatter(X[:, 0], X[:,1], s=100, c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label='Centroids')
plt.title('Iris Clusters and Centroids')
plt.xlabel('SepalLength')
plt.ylabel('SepalWidth')
plt.legend()

plt.show()
```
